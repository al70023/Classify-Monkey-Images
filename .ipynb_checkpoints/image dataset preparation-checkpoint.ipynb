```python
import torchvision
import torch
import torchvision.transforms as transforms
import os
import matplotlib.pyplot as plt
import numpy as np
os.environ['KMP_DUPLICATE_LIB_OK']='True'
```

```python
train_dataset_path = './monkeys/training/training'
test_dataset_path = './monkeys/validation/validation'
```

```python
mean = [0.4363, 0.4328, 0.3291]
std = [0.2129, 0.2075, 0.2038]



train_transforms = transforms.Compose([
    transforms.Resize((224,224)),  # experimental, in this case just 224x224 pixels, but model might perform differently
                                   # smaller images will help the NN train faster
                                   # many NN expect images to be same size
                                   # sometimes, images resized will help model to generalize better
                                   # make sure same Resize to calculate mean, std, as training 
    transforms.RandomHorizontalFlip(),  # also experimental, helps unbias the data and help expand view of NN
    transforms.RandomRotation(10), 
    transforms.ToTensor(),
    transforms.Normalize(torch.Tensor(mean), torch.Tensor(std)) # improves performance of model, generally always do it
])

test_transforms = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.ToTensor(),
    transforms.Normalize(torch.Tensor(mean), torch.Tensor(std))
])
```

```python
train_dataset = torchvision.datasets.ImageFolder(root=train_dataset_path, transform=train_transforms)
test_dataset = torchvision.datasets.ImageFolder(root=test_dataset_path, transform=test_transforms)
```

```python
def show_transformed_images(dataset):
    loader = torch.utils.data.DataLoader(dataset, batch_size=6, shuffle=True)
    batch = next(iter(loader))
    images, labels = batch
    
    grid = torchvision.utils.make_grid(images, nrow=3)
    plt.figure(figsize=(11,11))
    plt.imshow(np.transpose(grid, (1,2,0)))
    print('labels: ', labels)
```

```python
show_transformed_images(train_dataset)
```

```python
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=32, shuffle=False)
# start with batch size 32, then compare performance of NN at higher batch sizes
```
